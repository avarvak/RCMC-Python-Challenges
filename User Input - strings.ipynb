## <font color="blue"> Example 1: a string is like a list of characters </font>

In the code below, time is a string (that is, a text).

We can get the length of the string by using the function len().

The first character on the left has index 0, the second character on the left has index 1, etc.  We can read individual characters by specifying their index.  

One big difference between strings and lists is that we cannot change individual characters in a string.

```python
time = "10:34 am"
print("String length:", len(time))
print("First character:", time[0])
print("Second character:", time[1])
print("Third character:", time[3])
print("Fourth character:", time[4])
```

## Challenge 1: Check the last two characters

The code below asks the user to put in a time.  Check that the last two characters are either "am" or "pm".  If they are not, tell the user they didn't input the time properly.

```python
print("Input time, specifying am or pm, e.g.: 10:45 am or 2:40 pm")
time = input("Input time: ")

if ((time[len(time) - 2] == "a") or (time[len(time) - 2] == "p")) and (time[len(time) - 1] == "m"):
  print("You put in the time right.")
else:
  print("You put in the time wrong.")
```

## <font color="blue"> Example 2: user-defined function </font>

The code below defines a function that gets user input and returns a string composed of only the first and last character.

All the variables that are defined inside the function (indented) are local variables, which disappear after the function finishes its work.

The "return" command specifies the value to return.

```python
def firstLast():
    inp = input("Type anything: ")
    first = inp[0]
    last = inp[len(inp)-1]
    firstlast = first + last
    return firstlast

# call the function
result = firstLast()
print(result)
```
