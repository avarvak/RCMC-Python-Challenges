## Example 1

The code below has a hard-coded distance in feet and converts it into distance in miles.

```python
feet = 4321
miles = feet / 5280     # There are 5,280 feet in one mile
miles = round(miles, 2) # Rounds the result to two decimal digits
print(feet, " feet is ", miles, " miles")
```

## Example 3

We give the choice of which way to convert (feet to miles, or miles to feet) to the user.
The app assumes that the user would specify feet in whole number (integer), but can specify miles with fractional part (a real number, or "float" in Python).

```python
choice = input("Type fm for feet-to-miles, or mf for miles-to-feet: ")

if choice == 'fm':
    feet = int(input("Number of feet: "))
    miles = feet / 5280
    print(feet, " feet is ", round(miles,2), " miles")
elif choice == 'mf':
    miles = float(input("Number of miles: "))  # float is for real numbers, rather than integers
    feet = miles * 5280    # because miles is a float, here so is feet
    feet = int(round(feet, 0))  # round the feet to a whole number and convert it to integer
    print(round(miles, 2), " miles is ", feet, "feet")
else:
    print("I am confused by your choice.")
```
